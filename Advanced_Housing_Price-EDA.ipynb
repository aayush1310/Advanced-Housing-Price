# Missing Values

```python
df.isnull().sum()
```

```python
for cols in df.columns:
    print("Missing Values in ",cols,"===>",((df[cols].isnull().sum())/df[cols].shape[0])*100,"%")
```

```python
## Numerical Features using List Comprehension 
numerical_features = [features for features in df.columns if df[features].dtypes != 'O']
print("Total Number of Numerical Columns ", len(numerical_features))

df[numerical_features].head()
```

```python
# list of variables that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
print("Total Number of Temporal Features ", len(year_feature))
print("Lsit of Temporal features ",year_feature)
df[year_feature].head()
```

```python
## Categorical Features using List Comprehension
categorical_features = [features for features in df.columns if df[features].dtypes == 'O']
print("Total Number of Categorical Features ", len(categorical_features))

df[categorical_features].head()
```

### Missing Value in Continuous Variables

```python
numerical_features_with_na = [features for features in numerical_features if df[features].isnull().sum()>1]
```

```python
for cols in numerical_features_with_na:
    print("Missing Values in Column ",cols," ==> ",(df[cols].isnull().sum()/df[cols].shape[0])*100,"%")
```

### Missing Value in Categorical Variables

```python
categorical_features_with_na = [feature for feature in categorical_features if df[feature].isnull().sum()>1]
```

```python
for cols in categorical_features_with_na:
    print("Missing Values in Column ",cols," ==> ",(df[cols].isnull().sum()/df[cols].shape[0])*100,"%")
```
