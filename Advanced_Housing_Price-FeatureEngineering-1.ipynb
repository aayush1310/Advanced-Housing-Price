### Numerical Features

```python
numerical_features = [features for features in df.columns if df[features].dtypes != 'O']
```

### Categorical Features

```python
categorical_features = [features for features in df.columns if df[features].dtypes == 'O']
```

### Temporal Features

```python
temporal_features = [features for features in numerical_features if 'Yr' in features or 'Year' in features]
```

### Numerical Features with NaN

```python
numerical_features_with_na = [features for features in numerical_features if df[features].isnull().sum() > 1]
```

```python
categorical_features_with_na = [features for features in categorical_features if df[features].isnull().sum() > 1]
```

### Temporal Features with NaN

```python
temporal_features_with_na = [features for features in temporal_features if df[features].isnull().sum() > 1]
```

Since the data has many outliers, We Will handle the outliers using winsoriziation. We could have used median but here we
can see that median becomes 0. So, Winsorization is the best option

```python
from scipy.stats.mstats import winsorize
```

```python
data = df.copy()
```

```python
data = data.loc[data['MasVnrArea'].notnull() & data['LotFrontage']]
```

```python
data.head()
```

```python
for cols in numerical_features_with_na:
    sns.catplot(y=cols, data=data, kind='box')
```

```python
for cols in numerical_features_with_na:
    winsorize(data[cols], limits=[0.1, 0.2], inplace=True)
```

```python
mean_val = []
for cols in numerical_features_with_na:
    mean_val.append(data[cols].mean())
```

```python
for num, cols in enumerate(numerical_features_with_na, start=0):
    df[cols].fillna(mean_val[num], inplace=True)
```

```python
for cols in numerical_features_with_na:
    print("Missing Values in",cols,"==>",df[cols].isnull().sum())
```

```python
for cols in numerical_features_with_na:
    sns.catplot(y=cols, data=df, kind='box')
```

```python
for cols in numerical_features_with_na:
    winsorize(df[cols], limits=[0.1, 0.1], inplace=True)
```

```python
for cols in numerical_features_with_na:
    sns.catplot(y=cols, data=df, kind='box')
```

### Eliminiating Missing Values from Categorial Features

Replacing with most frequent values

```python
categorical_features_with_na
```

```python
for cols in categorical_features_with_na:
    print(df[cols].value_counts(dropna=False))
    print("\n")
```

```python
for cols in categorical_features_with_na:
    df[cols].fillna(df[cols].value_counts().index[0], inplace=True)
```

```python
for cols in categorical_features_with_na:
    print(cols," ",df[cols].isnull().sum())
```
